```r
c="Oligodendrocyte_precursor_cell"
```

```r
df1=NULL
#fn=paste0("/dfs3b/ruic20_lab/junw42/HCA_ON/data/7_DEG/genexp_donor_majorclass_final/snRNA_donor_num_cell_",c)
meta=read.table("/dfs3b/ruic20_lab/junw42/HCA_ON/data/7_DEG/genexp_sample_raw_atlas_meta_gsm_race_batch_sampleid",header=T,sep=",")

fn=paste0("/dfs3b/ruic20_lab/junw42/HCA_ON/data/7_DEG/snRNA_sample_majorclass_num_atlas_cell")
meta=meta[meta$tissue=="ON",]
data=read.table(fn,header=F,sep="\t")
#n=match(meta$sampleid,data$V1,nomatch=0)
data=data[data$V1%in%meta$sampleid,]
n=match(data$V1,meta$sampleid,nomatch=0)
data$age=meta[n,]$age_year
data$race=meta[n,]$race
data$gender=meta[n,]$gender
data=data[order(data$age),]
```

```r
sub=unique(data$V2)
```

```r
df=data %>% group_by(V1) %>% summarise(prop=V3/sum(V3),total=sum(V3),V3=V3)
df$ct=data$V2
df$age=data$age
df$race=data$race
df$gender=data$gender
```

```r
if(c=="Oligodendrocyte"){
#df=df[df$total>=500,]
df=df[(df$V3>=100)&(df$ct=="Oligodendrocyte"),]
#df=df[df$ct=="Oligodendrocyte",]

df$age_range="<60"
df[df$age>=60,]$age_range=">=60"

p=wilcox.test(df[df$age<55,]$prop, df[df$age>=60,]$prop,alternative="greater")$p.value
print(c)
print(p)

}
if(c=="Oligodendrocyte_precursor_cell"){
df=df[(df$V3>=20)&(df$ct=="Oligodendrocyte_precursor_cell"),]
#df=df[df$ct=="Oligodendrocyte_precursor_cell",]

df$age_range="<65"
df[df$age>=65,]$age_range=">=65"

p=wilcox.test(df[df$age<65,]$prop, df[df$age>=65,]$prop,alternative="greater")$p.value
print(c)
print(p)

}
#df$majorclass=c
#df1=rbind(df,df1)
mt=matrix(NA,ncol=2,nrow=length(sub))
#for(i in sub){
df1=df[df$ct==c,]
summary(lm(prop~age+race+gender,data=df1))
```
